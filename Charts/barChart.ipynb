```python
import matplotlib.pyplot as plt
```

### Bar charts

```python
products = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

# use plt.barh -- for horizontal display of data
plt.bar(products, sales, color="orange", label="Sales 2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product Sale Comparison")
plt.legend(loc="upper left")
plt.show()

plt.barh(products, sales, color="orange", label="Sales 2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product Sale Comparison")
plt.legend(loc="upper left")
plt.show()
```

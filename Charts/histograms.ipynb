```python
import matplotlib.pyplot as plt
```

### Histogram 

```python
## Syntax = plt.hist(data, bins=number_of_bins, color=colorname, edgecolor='black')
## bins decide in how many number of category we have to divide our data 


scores = [23,43,32,54,46,64,34,35,78,89,76,45,54,23,43,54]

plt.hist(scores, bins=5, color='pink', edgecolor='orange')
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Score Distribution of Students")
plt.show()

# height of bar is the number of values fall in that range 
# taller means more number 
```

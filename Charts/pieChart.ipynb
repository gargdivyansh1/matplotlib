```python
import matplotlib.pyplot as plt
```

### Pie Chart

```python
## Syntax = plt.pie(values, label=label_list , colors=colors_list, autopct='%1.1f%%')

regions = ["North", "South", "East", "West"]
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'skyblue', 'red', 'lightgreen'])
plt.title("Revenue Contribution By Region")
plt.show()
```

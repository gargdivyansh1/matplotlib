```python
import matplotlib.pyplot as plt
```

```python
## this is used for saving the data on the system 
## this is used for saving the chart in the different differnet extensions 

x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y, color='blue', marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.savefig("line_plot.png", dpi=300, bbox_inches='tight')
plt.savefig("line_plot.pdf", dpi=300, bbox_inches='tight')
plt.savefig("line_plot.jpg", dpi=300, bbox_inches='tight')
plt.savefig("line_plot.jpeg", dpi=300, bbox_inches='tight')
plt.show()
```

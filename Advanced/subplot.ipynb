```python
import matplotlib.pyplot as plt
```

### Subplot 

```python
x = [1,2,3,4]
y = [10,20,15,25]

## this is not the professional manner as we do not have the most command on the items 

plt.subplot(1,2,1) ## 1 row .. which will be containing 2 columns and i am using its first column 
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1, 2, 2) ## 1 row .. which is containing 2 columns and i am using 2 col this time
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

## here we will be using the object orient method 
## fig, ax = plt.subplots(nrows, ncols, figsize=(width , height))


## here we have decided that there will be 1 row and 2 columns and and the size of each fig
fig, ax = plt.subplots(1, 2, figsize = (10,5))

ax[0].plot(x,y)
ax[0].set_title("Line Graph")

ax[1].bar(x,y)
ax[1].set_title("Bar Graph")

plt.tight_layout()
plt.show()
```

### setting the title for whole plot box .. means for all the graphs


```python
x = [1,2,3,4]
y = [10,20,15,25]

fig, ax = plt.subplots(1, 2, figsize = (10,5))


## for setting the main title for all the graphs
fig.suptitle("Title for all the graphs")

ax[0].plot(x,y)
ax[0].set_title("Line Graph")

ax[1].bar(x,y)
ax[1].set_title("Bar Graph")

plt.tight_layout()
plt.show()
```

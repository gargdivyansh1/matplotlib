```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('MOCK_DATA.csv')
#print(df.head())

# for cleaning the data 
df.dropna(subset=['type', 'release_year_rating', 'rating', 'country', 'duration'])

type_counts = df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values , color=['skyblue', 'orange'])
plt.title('Number of Movies and TV shows on Netflix')
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```

```python
df = pd.read_csv('MOCK_DATA.csv')
df.dropna(subset=['type', 'release_year_rating', 'rating', 'country', 'duration'])

rating_counts = df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Content Rating')
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```

```python
movie_df = df[df['type'] == 'Movie'].copy()

## this is here we are converting the duration into minutes 
# movie_df['duration_int'] = movie_df['duration'].str.replace('min', '').astype(int)

plt.figure(figsize=(8,6))
plt.hist(movie_df['duration'], bins=30, color='purple', edgecolor='black')
plt.title("Distribution of Movie Duration")
plt.xlabel("Duration (minues)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

```python
release_count = df['release_year_rating'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_count.index , release_count.values, color='red', label='Data Points')

## here we are also constructin the line graph for showing the lines
plt.plot(release_count.index, release_count.values, color='blue', linestyle='-', linewidth=2, label='Trend Line')

plt.title("Release year vs numbre of showws")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
county_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(county_counts.index, county_counts.values, color='teal')
plt.title("Top 10 counties by number of shows")
plt.xlabel("Number of countries")
plt.ylabel("Country")
plt.tight_layout()
plt.show()



```

```python
content_by_year = df.groupby(['release_year_rating', 'type']).size().unstack().fillna(0)

fig, ax = plt.subplots(1,2, figsize=(12,5))

## first subplot
ax[0].plot(content_by_year.index, content_by_year['Movie'], color='blue')
ax[0].set_title("Movies Released Per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Number of Movies")

## Second subplot
ax[0].plot(content_by_year.index, content_by_year['TV show'], color='orange')
ax[0].set_title("TV shows Released Per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Number of TV shows")

fig.suptitle("Comparison of Movies and TV shows released Over year")

plt.tight_layout()
plt.show()
```

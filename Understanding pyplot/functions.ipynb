```python
import matplotlib.pyplot as plt
```

### plt.plot(), plt.xlabel(), plt.ylabel(), plt.title(), plt.legend(), plt.grid(), plt.xlim(), plt.ylim(), plt.xticks(), plt.yticks(), plt.show()

```python
x = ["Monday", "Tuesday", "Wednesday", "Thrusday", "Friday", "Saturday", "Sunday"]
y = [23,43,54,38,67,89,101]

plt.plot(x, y, color="blue", linestyle="--", linewidth=2, marker= 'o', label="Sales data")

plt.xlabel("Day of the week")

plt.ylabel("Sales on a day")

# 1. upper left 
# 2. lower right
plt.legend(loc="lower right")

plt.grid(color="gray", linestyle='--', linewidth=1)

plt.xlim("Monday", "Saturday")

plt.ylim(0, 100)

plt.xticks(["Monday", "Tuesday", "Wednesday", "Thrusday", "Friday", "Saturday", "Sunday"] , ["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7"])

plt.yticks([23,43,54,38,67,89,101], ["okay", "above avg", "good", "avg", "excellent", "fabulous", "extraordinary"])

plt.show()
```
